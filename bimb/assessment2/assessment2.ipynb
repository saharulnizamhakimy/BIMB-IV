```python
import requests
url = r"https://www.mudah.my/malaysia/properties-for-sale"
mudah = requests.get(url).content
Ads_Name = []
Ads_Datetime = []
Location = []
Price = []
properties_type = []
sqft = []

mudah
```

```python
from bs4 import BeautifulSoup
soup2 = BeautifulSoup(mudah)
soup2
```

```python
raw_data = soup2.find("div", {'class':"mw301 mw5"})
raw_data
```

```python
for rd in raw_data.findAll("div" ,{'class': "sc-kaNhvL biNyAq"}):
  print("-----------NEW RECORD------------")

  adsname = rd.findAll("a", { 'class' : "sc-gmeYpB cUDgyr"})
  if adsname:
        Ads_Name.append(adsname[0].find(string=True))
  else:
        Ads_Name.append(None)
  # Ads_Name.append(adsname[0].find(string=True))
  print(adsname[0].find(string=True))

  adsdttm = rd.findAll("span", { 'class' : "sc-qrIAp hgKLGX"})
  if adsdttm:
        Ads_Datetime.append(adsdttm[0].find(string=True))
  else:
        Ads_Datetime.append(None)
  # Ads_Datetime.append(adsdttm[0].find(string=True))
  print(adsdttm[0].find(string=True))

  loctn = rd.findAll("span", { 'title' : "Region"})
  if loctn:
        Location.append(loctn[0].find(string=True))
  else:
        Location.append(None)
  # Location.append(loctn[0].find(string=True))
  print(loctn[0].find(string=True))

  prc = rd.findAll("div", { 'class' : "sc-iSDuPN jwnHWZ"})
  if prc:
        Price.append(prc[0].find(string=True))
  else:
        Price.append(None)
  # Price.append(prc[0].find(string=True))
  print(prc[0].find(string=True))

  proptype = rd.findAll("div", { 'title' : "Category"})
  if proptype:
        properties_type.append(proptype[0].find(string=True))
  else:
        properties_type.append(None)
  # properties_type.append(proptype[0].find(string=True))
  print(proptype[0].find(string=True))

  sizsqft = rd.findAll("div", { 'title' : "Size"})
  if sizsqft:
        sqft.append(sizsqft[0].find(string=True))
  else:
        sqft.append(None)
  # sqft.append(sizsqft[0].find(string=True))
  print(sizsqft[0].find(string=True))
```

```python
import pandas as pd

cleanse_data = {
    "Ads_Name": Ads_Name,
    "Ads_Datetime": Ads_Datetime,
    "Location": Location,
    "Price": Price,
    "Properties_Type": properties_type,
    "Sqft": sqft
}

clean_data = pd.DataFrame(cleanse_data)
clean_data = clean_data[clean_data['Location']=='Johor']
clean_data = clean_data.head(100000)
```

```python
clean_data
```

```python
#fully referenced online
import pandas as pd
import os

output_dir = 'Output_Result'

os.makedirs(output_dir, exist_ok=True)

try:
    for ads_date, group_df in clean_data.groupby('Ads_Datetime'):
        partition_file_path = os.path.join(output_dir, f'Ads_Datetime={ads_date}.txt')
        group_df.to_csv(partition_file_path, index=False, sep='\t')
        print(f"Partition file saved successfully: '{partition_file_path}'")
except Exception as e:
    print(f"Error saving partitioned files: {e}")
```
